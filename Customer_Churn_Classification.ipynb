**Encoding multiple columns**

```python
car=[]
for i in df.columns:
  if df[i].dtype=='object':
    car.append(i)
    print(i)

from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for i in car:
  df[i]=df[[i]].apply(le.fit_transform)
df.head(3)
# df['Gender_Encoded']=le.fit_transform(df['gender'])
# df['Output_Encoded']=le.fit_transform(df['Churn'])
# df['P_Encoded']=le.fit_transform(df['Partner'])
# df['D_Encoded']=le.fit_transform(df['Dependents'])
# df['PS_Encoded']=le.fit_transform(df['PhoneService'])
# df['PB_Encoded']=le.fit_transform(df['PaperlessBilling'])
# df['ONS_Encoded']=le.fit_transform(df['OnlineSecurity'])
```

```python
x=df.drop(['customerID','Churn'],axis=1)
y=df['Churn']
```

```python
x.columns
```

```python
y.head(3)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=10)
knn.fit(x_train,y_train)
```

```python
y_pred=knn.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
accuracy_score(y_test,y_pred)
# confusion_matrix(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```
